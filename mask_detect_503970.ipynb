```
def load_image(path):
    img =tf.io.read_file(path)
    img=tf.image.decode_jpeg(img,channels=3)
    img=tf.image.resize(img,[360,360])
    #img=tf.image.random_crop(img,[224,224,3])
    #img=tf.image.random_flip_left_right(img)
    #img =tf.image.random_brightness(img,max_delta=0.5)
    #img =tf.image.random_contrast(img,0,0.5)
    #img=tf.image.resize(img,[224,224])
    img=img/127.5-1 #归一化
    return img
```

```
#%tensorflow_version 2.x
import tensorflow as tf
img_ds=tf.data.Dataset.from_tensor_slices(train_img)
```

```
img_ds =img_ds.map(load_image)
```

```
lable_ds =tf.data.Dataset.from_tensor_slices((out_1,out_2,out_3,out_4,index))
```

```
dataset =tf.data.Dataset.zip((img_ds,lable_ds))
```

```
BATCH_SIZE =32
```

```
dataset =dataset.shuffle(len(train_img)).batch(BATCH_SIZE).repeat()
```

```
net =tf.keras.applications.Xception(weights='imagenet',
                                         include_top=False,
                                        input_shape =(None,None,3)
                                        )
```

```
inputs =tf.keras.layers.Input(shape=(None,None,3))
#函数式API
x =net(inputs)
x =tf.keras.layers.GlobalAveragePooling2D()(x)
x1 =tf.keras.layers.Dense(2048,activation='relu')(x)
x1 =tf.keras.layers.Dense(512,activation='relu')(x1)
x1 =tf.keras.layers.Dense(256,activation='relu')(x1)


#out =tf.keras.layers.Dense(4,name='out')(x1)
out_1=tf.keras.layers.Dense(1,name='out1')(x1)
out_2=tf.keras.layers.Dense(1,name='out2')(x1)
out_3=tf.keras.layers.Dense(1,name='out3')(x1)
out_4=tf.keras.layers.Dense(1,name='out4')(x1)
###########################################
x2 =tf.keras.layers.Dense(512,activation='relu')(x)
x2 =tf.keras.layers.Dense(256,activation='relu')(x1)
index=tf.keras.layers.Dense(1,name='index',activation='sigmoid')(x2)

prediction =[out_1,out_2,out_3,out_4,index]
#prediction =[index,classify]


model =tf.keras.models.Model(inputs=inputs,outputs=prediction)
```

```
model.compile(
    optimizer=tf.keras.optimizers.Adam(0.0001),
    loss ={'out1':'mse',
           'out2':'mse',
           'out3':'mse',
           'out4':'mse',
           'index':'binary_crossentropy'},
    # loss={'index':'binary_crossentropy' ,
    #    'classify':'sparse_categorical_crossentropy'},
    metrics=['mse']
)
```
